```python
drop_column = ['url','address','phone','reviews_list']
for i in drop_column:
    df.drop(i,axis=1,inplace=True)
```

```python
df.isna().sum()
```

<font color='green'>Firstly we will delete all 21 rows where location is NUll & approx_cost column also, it makes no sense to assume that</font>

```python
df['location'].dropna(how=any,axis=0,inplace=True)
```

```python
df['approx_cost(for two people)'].dropna(how=any,axis=0,inplace=True)
```

<font color='green'>We can see that rate column has a lot of rows as NaN ,  basically either a rating was not given or it was not recorded .  We will transform all Null values to 'Not Reported' so as to not lose any information</font>

```python
df.loc[df['rate'].isna(),'rate'] = 'Not Rated'
```

```python
df['rest_type'].value_counts()
```

<font color='green'>Since we dont know the the Type of Restaurant for and the number of records for the same is also low we shall remove the rows with missing values

```python
df['rest_type'].dropna(how='any',axis=0,inplace=True)
```

<font color='green'>Now, we Can see that a restaurant can have two types together as comma seperated, since this comma seperated types is not of much use to us as during visualization it will be seen as a unique entity , we shall divide it , the approach we shall use is similar to one hot encoding . i.e for each unique type of rest we will have a column, then for every row we shall check the type and whatever type is mentioned be it single or comma seperated value , ww will give a values of 1 in the column for that type

```python
unique_rest_type = set()

for rest in df['rest_type']:
    rest_str = str(rest)
    unique_rest_type.update(rest_str.split(', '))
```

```python
unique_rest_type
```

<font color='green'> Now that we have all the unique values of rest_type we will create the columns for them

```python
for rest in unique_rest_type:
    df['rest_type_'+rest]=0
```

```python
df.head(2)
```

<font color='green'> Now we shall enumerate over all the rows get the row id and the rest_type and based on the value of rest_type assign 1 to the respective columns

```python
for i,rest in enumerate(df['rest_type']):
    rest_str = str(rest)
    for x in rest_str.split(', '):
        df.at[i,'rest_type_'+x]=1
```

```python
df[df['rest_type_Dhaba']==1]
```

<font color = 'green'>In this example we can see that when rest_type == Dhaba our new column rest_typeClub has a value of 1, GREAT GOOD JOB!! One Step Closer  
    
<font color = 'green'>Moving forward to dish_liked 

```python
df['dish_liked'].isna().sum()
```

<font color = 'green'> With around 28,000 rows of Null value it is important to look at this , basically we can assume that no dish was liked at the restuarant , so we shall convert all Null values to 'None'

```python
df['dish_liked'].fillna('None',inplace=True)
```

```python
df['dish_liked'].value_counts()
```

<font color='green'> With this column the same problem arises of dishes being mentioned as comma seperated so we shall perform the same task of creating diff columns for each dish as a whole

```python
unique_dishes = set ()

for i in df['dish_liked']:
    dish_str = str(i)
    unique_dishes.update(dish_str.split(', '))
```

```python
len(unique_dishes)
```

```python
unique_dishes
```

<font color='Green'>It makes absolutely no sense to compute additional 2793 rows for visualization , so we shall create a temporary df in which we will create these one hot encoding columns and we will sort them based on count and take the number of columns that make most sense

```python
df_dishes = df['dish_liked'].str.get_dummies(sep=',')
```

```python
df_dishes
```

<Font color='green'>Now we have the columns ready and we shall pick the top 51 columns where the sum is largest which basically means the the dishes which were ordered the most and concat to our main df , further we should remove the joint dish_liked column from our df 

```python
top_dishes = df_dishes.sum().nlargest(51)
df_top_dishes = df_dishes[top_dishes.index]
df = pd.concat([df, df_top_dishes], axis=1)

# Drop the original "dish_liked" column if needed
df.drop('dish_liked', axis=1, inplace=True)
df.drop('None', axis=1, inplace=True)
```

```python
df.drop('cuisines', axis=1, inplace=True)
df.drop('menu_item', axis=1, inplace=True)
```

```python
df.to_csv('zomato_transform.csv',index=False)
```
